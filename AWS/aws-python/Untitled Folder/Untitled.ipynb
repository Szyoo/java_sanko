```python
import boto3
import json
import sys
from PIL import Image

TargetImage = input('顔の検出をする画像ファイル名を指定してください(.jpgまで)')
    
rekognition = boto3.client('rekognition')
with open(TargetImage , 'rb' ) as file:
    result = rekognition.detect_faces(Image={'Bytes' : file.read()})
    print(json.dumps(result, indent=4))
    
    
image_in = Image.open(TargetImage)
w , h = image_in.size
image_out = Image.new('RGB' , (w , h) , (200, 200, 200))
for face in result['FaceDetails']:
    box = face['BoundingBox']
    left = int(box['Left'] * w)
    top = int(box['Top'] * h)
    right = left + int (box['Width'] * w )
    bottom = top + int (box['Height'] * h )
    image_out.paste(
        image_in.crop((left, top, right, bottom)),
        (left, top))
image_out.save('show_question_' + TargetImage)
image_out.show()
```

```python
import boto3
import json
import sys
from PIL import Image
import glob
import os
import shutil


# TestImage
# pathname = input('顔の検出をする画像があるフォルダーのパスー名を指定してください:')
pathname = "TestImage"
logLevel = "verbose"
# logLevel = "normal"

rekognition = boto3.client('rekognition')
dict_multiple = {}  # 多人路径字典，键值对为<path, 人头数>
count1 = 0  # source循环计数器
count2 = 0  # target循环计数器
personCount = 0  # 识别到的个人文件夹数量
matchCount = 0  # 单次循环内匹配成功的target次数
totalMatch = 0  # 共计匹配成功的target次数
list_delete = []  # 待删除列表，用于存放符合条件的path，每次循环刷新

paths = glob.glob(f'{pathname}\**\*', recursive=True)


def log(content, level):
    global logLevel
    if not isinstance(content, str):
        content = str(content)
    if level == "normal" or logLevel == level:
        if level == "verbose":
            content = "    "+content
        print(content)


try:
    os.makedirs(pathname+"/error")  # 新建Error文件夹
except:
    log("error文件夹已存在", "verbose")
else:
    log("error文件创建成功", "verbose")

totalPathNumber = paths.__len__()


def openSource(sourcePath):
    # 声明全局变量，否则会将此变量视为本地变量而导致报错
    global rekognition
    global dict_multiple
    global list_delete
    global personCount
    global matchCount
    global count1
    global count2
    global paths
    global totalPathNumber
    count2 = 0  # 重置target计数器
    matchCount = 0  # 重置match计数器
    try:
        with open(sourcePath, 'rb') as source:  # 打开source图片
            detectResult = rekognition.detect_faces(Image={'Bytes': source.read()})  # 检测是否存在人脸
            number = len(detectResult['FaceDetails'])  # 检测到的人脸数量
            if number == 1:  # source检测到单人
                list_delete.append(sourcePath)  # 将此图片记为待删除
                savePath = sourcePath.replace(".JPG", "")
                os.makedirs(savePath)  # 以sourcePath新建个人文件夹
                personCount += 1
                shutil.copy(sourcePath, savePath)  # 将该文件复制到个人文件夹
                log("当前进度："+str(count1)+"/"+str(totalPathNumber)+"，检测到source图片，路径为" + sourcePath+"，开始匹配...", "normal")  # 打印当前source路径
                for targetPath in paths:  # 循环文件夹下的每一个图片作为本次的target
                    source.seek(0)  # 在每次循环中重置source文件的指针，以避免读取不到内容
                    if targetPath != sourcePath:  # 避免重复读取同一个source文件
                        count2 += 1  # 循环计数器+1
                        openTarget(source, targetPath, savePath)
                log("    当前共检测到"+str(matchCount)+"张匹配图片，已全部移动到指定个人文件夹", "normal")
            else:  # source检测到多人
                log("当前进度："+str(count1)+"/"+str(totalPathNumber)+"，检测到多人source图片，路径为" + sourcePath+"，人数为："+str(number)+"记录并跳过", "normal")  # 打印当前source路径
                dict_multiple[sourcePath] = number  # 以<路径，人数>的方式记录到字典中
    except (PermissionError, FileNotFoundError):
        log("当前进度："+str(count1)+"/"+str(totalPathNumber)+"，当前文件已移动，跳过，文件路径："+sourcePath, "verbose")
    except:
        print("Unexpected error:" + sys.exc_info()[0])
        raise


def openTarget(source, targetPath, savePath):
    # 声明全局变量，否则会将此变量视为本地变量而导致报错
    global count2
    try:
        with open(targetPath, 'rb') as target:  # 打开target文件
            compare(source, target, targetPath, savePath)
    except (PermissionError, FileNotFoundError):
        log("当前进度："+str(count2)+"/"+str(totalPathNumber)+"，当前文件已移动，跳过，文件路径："+targetPath, "verbose")
    except:
        print("Unexpected error:" + sys.exc_info()[0])
        raise


def recordUnmatchedFace(compareResult, targetPath):
    # 声明全局变量，否则会将此变量视为本地变量而导致报错
    global dict_multiple
    global list_delete
    if targetPath not in dict_multiple:  # 如果此时该target不在已识别到的多人图片dict中
        number = len(compareResult['UnmatchedFaces'])  # 检测到的人脸数量
        if number == 0:
            shutil.copy(targetPath, "error")  # 将该文件复制到错误文件夹
            list_delete.append(targetPath)
        else:
            dict_multiple[targetPath] = number  # 以<路径，人数>的方式记录到字典中
    else:  # 如果该target是在已识别到的多人图片dict中已经记录过的
        number = dict_multiple[targetPath]  # 获取记录中的人数
        dict_multiple[targetPath] = number-1  # 将该人数减1，代表图片中剩余待识别人数


def compare(source, target, targetPath, savePath):
    # 声明全局变量，否则会将此变量视为本地变量而导致报错
    global rekognition
    global dict_multiple
    global list_delete
    global count2
    global matchCount
    try:
        compareResult = rekognition.compare_faces(  # 进行compare操作
            SourceImage={'Bytes': source.read()},
            TargetImage={'Bytes': target.read()})
    except rekognition.exceptions.InvalidParameterException as e:
        log(str(count2)+".出现错误，该图片识别不到任何人物，已移动至error文件夹，当前Target文件路径为："+targetPath, "verbose")
        shutil.copy(targetPath, pathname+"/error")  # 将该文件复制到错误文件夹
        list_delete.append(targetPath)
    except:
        print("Unexpected error:" + sys.exc_info()[0])
        raise
    else:
        if len(compareResult["FaceMatches"]) > 0:  # 如果找到了相似的脸
            similarity = compareResult["FaceMatches"][0]["Similarity"]
            if similarity > 96:  # 在对比结果中FaceMatches的list中获取Similarity，如果大于95则认为是识别成功
                matchCount = matchCount+1
                shutil.copy(targetPath, savePath)  # 将该文件复制到个人文件夹
                if len(compareResult["UnmatchedFaces"]) > 1:  # 如果在对比结果中存在识别失败的人脸，代表source里不止一个人
                    log(str(count2)+".识别到包含指定人物的多人图片，复制到个人文件夹，路径为"+targetPath + "，相似度为：" + str(similarity), "verbose")  # 打印相似度
                    recordUnmatchedFace(compareResult, targetPath)
                else:  # 不存在识别失败的人脸，代表source是单人的，则记录到待删除list中
                    log(str(count2)+".识别到指定人物单人图片，复制到个人文件夹，并记录为待删除，路径为"+targetPath + "，相似度为：" + str(similarity), "verbose")  # 打印相似度
                    list_delete.append(targetPath)
            else:
                log(str(count2)+".匹配到低相似度图片，跳过，路径为"+targetPath + "，相似度为：" + str(similarity), "verbose")
        elif len(compareResult["UnmatchedFaces"]) > 1:  # 如果在对比结果中存在识别失败的人脸，代表source里不止一个人
            if targetPath not in dict_multiple:
                recordUnmatchedFace(compareResult, targetPath)
                log(str(count2)+".未识别到指定人物，且为未记录的多人图片，记录并跳过，路径为"+targetPath, "verbose")
            else:
                log(str(count2)+".未识别到指定人物，且为已记录的多人图片，跳过，路径为"+targetPath, "verbose")
        else:
            log(str(count2)+".未识别到指定人物，且为单人图片，跳过，unmatch："+str(len(compareResult["UnmatchedFaces"]))+"，路径："+targetPath, "verbose")


log("共检测到"+str(totalPathNumber)+"个文件，开始人物识别分类流程", "normal")
for sourcePath in paths:  # 循环文件夹下的每一个图片作为本次的source
    count1 += 1  # 循环计数器+1
    if sourcePath not in dict_multiple:  # 如果该图片不属于已检测过的多人的图片，即为单人图片，或者未检测过的多人图片
        openSource(sourcePath)
    else:
        log("当前进度："+str(count1)+"/"+str(totalPathNumber)+"，检测到已存在的多人source图片，路径为" + sourcePath+"，记录并跳过", "normal")  # 打印当前source路径
    for path in list_delete:  # 在一次source循环中，最后将本次待删除list中的文件都删除
        os.remove(path)  # 删除列表中文件
        log("删除已标记单人文件，路径为："+path, "verbose")
    list_delete.clear()  # 清空该列表
for path in dict_multiple:
    # 全部source循环结束后，删除多人图片dict中剩余待识别人数为0的图片，留下仍有未识别人脸的照片
    if dict_multiple[path] == 0:
        os.remove(path)
        log("删除已全部识别完的多人文件，路径为："+path, "verbose")
log("完成，共成功识别并分类"+str(personCount)+"人，剩余未被移动到个人文件夹的照片，均为存在未能成功分类的人物的图片", "normal")
```
